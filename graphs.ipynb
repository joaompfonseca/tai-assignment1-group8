```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('results/window.csv')

plt.plot(data['window'], data['time'])
plt.xlabel('Window Size')
plt.ylabel('Time (s)')
plt.title('Window Size vs Time')
plt.show()
```

```python
min_bps_index = np.argmin(data['bps']) 
min_bps_value = data['bps'][min_bps_index] 
min_window = data['window'][min_bps_index] 

plt.plot(data['window'], data['bps'])
plt.xlabel('Window Size')
plt.ylabel('Bits per Symbol')
plt.title('Window Size vs Bits per Symbol')

plt.scatter(min_window, min_bps_value, color='red')

plt.show()
```

```python
data = pd.read_csv('results/lookback_size.csv')

plt.plot(data['lookback_size'], data['time'])
plt.xlabel('Lookback Size')
plt.ylabel('Time (s)')
plt.title('Lookback Size vs Time')
plt.show()
```

```python
min_bps_index = np.argmin(data['bps']) # Find the index of the minimum value in 'bps'
min_bps_value = data['bps'][min_bps_index] # Get the minimum value
min_lookback_size = data['lookback_size'][min_bps_index] # Get the corresponding 'lookback_size'

# Plot the original data
plt.plot(data['lookback_size'], data['bps'])
plt.xlabel('Lookback Size')
plt.ylabel('Bits per Symbol')
plt.title('Lookback Size vs Bits per Symbol')

# Add a red dot at the minimum value
plt.scatter(min_lookback_size, min_bps_value, color='red')

# Show the plot
plt.show()
```
