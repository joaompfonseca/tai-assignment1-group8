```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('results/window.csv')

plt.plot(data['window'], data['time'])
plt.xlabel('Window Size')
plt.ylabel('Time (s)')
plt.title('Window Size vs Time')
plt.show()
```

```python
plt.plot(data['window'], data['bps'])
plt.xlabel('Window Size')
plt.ylabel('Bits per Second')
plt.title('Window Size vs Bits per Second')
plt.show()
```

```python
data = pd.read_csv('results/original_comp.csv')

x = ["pred","gzip", "bzip2", "tarxz"]
y = [data["pred"][0],data["gzip"][0], data["bzip2"][0], data["tarxz"][0]]

plt.bar(x, y)
plt.xlabel('Compression Method')
plt.ylabel('Rate of Compression')
plt.title('Compression Method vs Rate of Compression (Original File)')
plt.show()
```

```python
data = pd.read_csv('results/compare.csv')

x = ["original", "mutate_0.1", "mutate_0.2", "mutate_0.3", "mutate_0.4", "mutate_0.5", "mutate_0.6", "mutate_0.7", "mutate_0.8", "mutate_0.9"]

y = [data["pred"][0], data["mutate_0.1"][0], data["mutate_0.2"][0], data["mutate_0.3"][0], data["mutate_0.4"][0], data["mutate_0.5"][0], data["mutate_0.6"][0], data["mutate_0.7"][0], data["mutate_0.8"][0], data["mutate_0.9"][0]]

plt.figure(figsize=(12,5))
plt.bar(x, y)
plt.xlabel('Files')
plt.ylabel('Rate of Compression')
plt.title('Files vs Rate of Compression')
plt.show()
```
